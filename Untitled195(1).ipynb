# Seaborn

Seaborn is a visualization library built using matplotlib that
focuses on creating standard statistical plots with a sample one
line function call

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\SRKT\Downloads\heart-disease-dataset\heart.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(8,4))
sns.distplot(df['age'],kde=False,bins=30,color='red') #distribution of data
```

```python
df.head()
```

```python
sns.countplot(df['sex'])
```

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot('cp',data=df,hue='sex')
```

```python
sns.boxplot(x='sex',y='age',data=df,hue='sex') # x=categorical features & y=continuos features
```

```python
sns.scatterplot(x='chol',y='trestbps',data=df,hue='sex')
```

```python
sns.pairplot(df)
```
